# Установка и загрузка пакетов

```r

install.packages("dplyr")
install.packages("ggplot2")
library(dplyr)
library(ggplot2)
```

# Генерация датасета

```r

set.seed(42)
n <- 5000

dataset <- data.frame(
  order_id = 1:n,
  customer_id = sample(1:1000, n, replace = TRUE),
  order_date = as.Date("2023-01-01") + sample(0:365, n, replace = TRUE),
  product_id = sample(1:200, n, replace = TRUE),
  category = sample(c("Электроника", "Одежда", "Продукты", "Косметика", "Книги"), n, replace = TRUE),
  quantity = sample(1:10, n, replace = TRUE),
  total_amount = round(runif(n, 10, 500), 2),
  discount = round(runif(n, 0, 0.5), 2)
)

write.csv(dataset, "orders.csv", row.names = FALSE)
```

# 1. Общая сумма продаж по категориям

```r

dataset %>%
  group_by(category) %>%
  summarise(total_sales = sum(total_amount))
```

# 2. Анализ количества заказов по месяцам

```r
dataset %>%
  mutate(month = as.Date(paste0(format(order_date, "%Y-%m"), "-01"))) %>%  # Преобразуем в дату
  group_by(month) %>%
  summarise(order_count = n(), .groups = 'drop') %>%  # Убедимся, что группировка сброшена
  ggplot(aes(x = month, y = order_count)) +
  geom_line() +  # Теперь линия будет построена корректно
  theme_minimal()
```

# 3. Топ-5 самых продаваемых товаров

```r

dataset %>%
  group_by(product_id) %>%
  summarise(total_quantity = sum(quantity)) %>%
  arrange(desc(total_quantity)) %>%
  head(5)
```

# 4. Средний чек по месяцам

```r

dataset %>%
  mutate(month = format(order_date, "%Y-%m")) %>%
  group_by(month) %>%
  summarise(avg_check = mean(total_amount))
```

# 5. Гистограмма количества заказов по дням недели

```r

dataset %>%
  mutate(weekday = weekdays(order_date)) %>%
  ggplot(aes(x = weekday)) + geom_bar() + theme_minimal()
```

# 6. Корреляция между скидками и суммой заказов

```r

cor(dataset$discount, dataset$total_amount)
```

# 7. Анализ среднего количества товаров в заказе

```r

dataset %>%
  group_by(order_id) %>%
  summarise(avg_items = mean(quantity)) %>%
  summarise(overall_avg = mean(avg_items))
```

# 8. Доля заказов со скидками

```r

dataset %>%
  summarise(discounted_orders = mean(discount > 0) * 100)
```

# 9. Аномальные заказы (сумма выше 3 стандартных отклонений от среднего)

```r

threshold <- mean(dataset$total_amount) + 3 * sd(dataset$total_amount)
dataset %>% filter(total_amount > threshold)
```

# 10. Распределение скидок по категориям

```r

dataset %>%
  group_by(category) %>%
  summarise(avg_discount = mean(discount)) %>%
  ggplot(aes(x = category, y = avg_discount)) + geom_col() + theme_minimal()
```
