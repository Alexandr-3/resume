```python
import pandas as pd
import numpy as np
from faker import Faker
import matplotlib.pyplot as plt

# Инициализация генератора случайных данных
fake = Faker()
np.random.seed(42)

# Генерация данных
num_orders = 5000  # Количество заказов

order_ids = np.arange(1, num_orders + 1)
customer_ids = np.random.randint(1, 1000, num_orders)
order_dates = [fake.date_time_between(start_date="-1y", end_date="now") for _ in range(num_orders)]
product_ids = np.random.randint(1, 200, num_orders)
categories = np.random.choice(["Электроника", "Одежда", "Продукты", "Косметика", "Книги"], num_orders)
quantities = np.random.randint(1, 10, num_orders)
total_amounts = np.round(np.random.uniform(10, 500, num_orders), 2)
discounts = np.round(np.random.uniform(0, 0.5, num_orders), 2)

# Создание DataFrame
df = pd.DataFrame({
    "order_id": order_ids,
    "customer_id": customer_ids,
    "order_date": order_dates,
    "product_id": product_ids,
    "category": categories,
    "quantity": quantities,
    "total_amount": total_amounts,
    "discount": discounts
})

# Сохранение в CSV
df.to_csv("orders.csv", index=False)

print("Датасет orders.csv успешно сгенерирован!")
```

Загружаем единый датасет
Ожидаемые столбцы: order_id, customer_id, order_date, product_id, category, quantity, total_amount, discount


```python
orders = pd.read_csv("orders.csv")
orders["order_date"] = pd.to_datetime(orders["order_date"])
```

1. Анализ продаж по категориям товаров
Задача: Посчитать сумму продаж по категориям и построить столбчатую диаграмму

```python
category_sales = orders.groupby("category")["total_amount"].sum()
category_sales.plot(kind="bar", title="Сумма продаж по категориям")
plt.show()
```

2. Анализ оттока клиентов
Задача: Найти клиентов, которые сделали только один заказ

```python
churn_customers = orders["customer_id"].value_counts()
churn_customers = churn_customers[churn_customers == 1]
print(f"Число ушедших клиентов: {len(churn_customers)}")
```

3. Средняя выручка по месяцам
Задача: Найти среднюю выручку по месяцам

```python
monthly_avg_revenue = orders.groupby(orders["order_date"].dt.month)["total_amount"].mean()
monthly_avg_revenue.plot(kind="line", title="Средняя выручка по месяцам")
plt.show()
```

4. Выявление аномальных транзакций
Задача: Найти заказы выше 3 стандартных отклонений от среднего

```python
mean, std = orders["total_amount"].mean(), orders["total_amount"].std()
anomalies = orders[orders["total_amount"] > mean + 3 * std]
print(anomalies)
```

5. Анализ популярности товаров
Задача: Найти топ-5 самых популярных товаров по количеству проданных единиц


```python
top_products = orders.groupby("product_id")["quantity"].sum().nlargest(5)
top_products.plot(kind="bar", title="Топ-5 товаров по продажам")
plt.show()
```

6. Распределение заказов по часам
Задача: Построить гистограмму количества заказов по часам


```python
orders["hour"] = orders["order_date"].dt.hour
orders["hour"].hist(bins=24)
plt.title("Распределение заказов по часам")
plt.show()
```

7. Средний чек по месяцам
Задача: Найти средний чек за каждый месяц

```python
avg_check = orders.groupby(orders["order_date"].dt.month)["total_amount"].mean()
avg_check.plot(kind="line", title="Средний чек по месяцам")
plt.show()
```

8. Доля активных пользователей
Задача: Посчитать процент клиентов, совершивших заказ за последние 6 месяцев


```python
active_customers = orders[orders["order_date"] >= "2024-01-01"]["customer_id"].nunique()
total_customers = orders["customer_id"].nunique()
print(f"Доля активных клиентов: {active_customers / total_customers * 100:.2f}%")
```

9. Анализ категорий товаров по выручке
Задача: Найти категории с наибольшей выручкой

```python
category_revenue = orders.groupby("category")["total_amount"].sum()
category_revenue.plot(kind="bar", title="Выручка по категориям")
plt.show()
```

10. Корреляция между скидками и продажами
Задача: Проверить, есть ли связь между скидками и объемами продаж

```python
correlation = orders[["discount", "quantity"]].corr()
print("Корреляция между скидками и объемом продаж:")
print(correlation)
```
